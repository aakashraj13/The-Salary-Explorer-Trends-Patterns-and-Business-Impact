## ***3.Data Wrangling***


```python
# Write your code to make your dataset analysis ready. 

df.head(4) # Displaying first few rows of the dataset

# Drop Unnamed column if present
df = df.loc[:, ~df.columns.str.contains('unnamed',case = False)]



# Check for missing values
missing_values = df.isnull().sum() 

# Drop rows with too many missing values
df = df.dropna(subset = ["Job Title","Salary Estimate", "Job Description"]) 

# Remove duplicate rows
df = df.drop_duplicates()

# Extract numeric salary information

df = df[df['Salary Estimate'] != '-1']  # Remove rows with missing salary
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split('(')[0]) # Remove extra text(Glassdoor est.) 
df['Salary Estimate'] = df['Salary Estimate'].str.replace('K','').str.replace('$','').str.replace('Employer Provided Salary:','').str.replace('Per Hour','')  # Remove symbols and text 
df['Salary Min'] = df['Salary Estimate'].apply(lambda x: int(x.split('-')[0])) #splitting minimun salary 
df['Salary Max'] = df['Salary Estimate'].apply(lambda x: int(x.split('-')[1]) if '-' in x else int(x)) # splitting maximun salary
df['Average Salary'] = (df['Salary Min'] + df['Salary Max'])/2  #Finding average salary 


# Standardize Company Names (remove ratings) 
df['Company Name'] = df['Company Name'].apply(lambda x: x.split('\n')[0])

# Extract location details
df[['City', 'State']] = df['Location'].str.split(',', expand = True, n=1)



# Convert categorical variables
df['Type of ownership']= df['Type of ownership'].astype('category')
df['Industry'] = df['Industry'].astype('category')
df['Sector'] = df['Sector'].astype('category')
df['Sector'] = df['Sector'].replace('', 'unknown')

# replacing -1 to blank space
df['Competitors'] = df['Competitors'].str.replace('-1','') 
# Replace '-1' with NaN
df['Founded'] = df['Founded'].replace([-1],np.nan) 
# Drop rows where 'Founded' is now NaN
df = df.dropna(subset = ['Founded'])
df['Founded'] = df['Founded'].astype('int')
df[['Industry','Sector']] = df[['Industry','Sector']].replace('-1','')

#Replacing text in state column
df['State'] = df['State'].str.replace('Los Angeles,','')
#Replacing text in size column
df['Size'] = df['Size'].str.replace('to','-')
df['Size'] = df['Size'].astype('category')



# Replace '-1' and 'Unknown' with NaN
df['Size'] = df['Size'].replace(['-1', 'Unknown'], np.nan)
# Drop rows where 'Size' is now NaN
df = df.dropna(subset=['Size'])




# Replace '-1' with NaN
df['Rating'] = df['Rating'].replace(-1, np.nan)
# Drop rows where 'Rating' is now NaN
df = df.dropna(subset=['Rating'])

#Replace '-1' and 'Unknown' with NaN
df['Type of ownership'] = df['Type of ownership'].replace(['-1', 'Unknown'], np.nan)
# Drop rows where 'Size' is now NaN
df = df.dropna(subset=['Type of ownership'])

#extract numeric revenue information
df = df[df['Revenue']!= '-1']  # Remove rows with missing revenue
df['Revenue'] = df['Revenue'].apply(lambda x: x.split('(')[0]) # Remove extra text(USD) 
df['Revenue'] = df['Revenue'].str.replace('$','').str.replace('to','-').str.replace('million','').str.replace('billion','').str.replace('+','').str.replace('Less than','')
df = df[df['Revenue']!= 'Unknown / Non-Applicable']  
df.head(4)
```

### What all manipulations have you done and insights you found?

1. Dropped Unnamed Columns
- Removed unnecessary unnamed columns that might have been generated during data extraction.
2. Handled Missing Values
- Checked for missing values in key columns like Job Title, Salary Estimate, and Job Description.
- Dropped rows with too many missing values.
3. Removed Duplicates
- Ensured data integrity by removing duplicate rows.
4. Extracted Numeric Salary Information
- Removed non-numeric text (e.g., "Glassdoor est.") from salary estimates.
- Converted salary values into Salary Min, Salary Max, and calculated Average Salary.
5. Standardized Company Names
- Removed ratings from company names to keep them clean.
6. Extracted Location Details
- Split Location into City and State for better geographic analysis.
7. Converted Categorical Variables
- Converted Type of ownership, Industry, Sector, and Size to categorical data types for efficient analysis.
8. Cleaned and Replaced Missing Values
- Replaced -1 and Unknown values in Competitors, Founded, Size, Type of Ownership, and Revenue with NaN.
- Dropped rows where Founded, Size, and Revenue were unknown or missing.

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

### Chart 1 - Top 10 Most Common Jobs and Their Average Salary

```python
# Find top 10 most common jobs and their average salary

# Find top 10 most common jobs
top_jobs = df['Job Title'].value_counts().head(10).index

# Calculate average salary for these jobs
top_jobs_salary = df[df['Job Title'].isin(top_jobs)].groupby('Job Title')['Average Salary'].mean().dropna()

```

```python
#import libraries 
import matplotlib.pyplot as plt 
import seaborn as sns

# Set Seaborn style and color palette
sns.set_style("darkgrid")  # Options: white, dark, whitegrid, darkgrid, ticks
sns.set_palette("rocket")   # Options: deep, muted, bright, pastel, dark, colorblind

# Set figure size
plt.figure(figsize=(12, 6))

# Sort jobs by salary for better visualization
top_jobs_salary = top_jobs_salary.sort_values(ascending=True)

# Create a horizontal bar plot
ax = sns.barplot(
    x=top_jobs_salary.values, 
    y=top_jobs_salary.index, 
    palette="ch:s=-.2,r=.6", 
    edgecolor="black",
    hue = top_jobs_salary.index
)

# Annotate bars with salary values
for p in ax.patches:
    ax.annotate(
        f"${p.get_width():,.0f}",  # Format salary values
        (p.get_width(), p.get_y() + p.get_height() / 2),
        ha='left', 
        va='center',
        fontsize=10, 
        fontweight='bold', 
        color='black'
    )

# Labels and title
plt.xlabel("Average Salary ($)", fontsize=12, fontweight='bold', labelpad=10)
plt.ylabel("Job Title", fontsize=12, fontweight='bold', labelpad=10)
plt.title("Top 10 Most Common Job Titles and Their Average Salaries", fontsize=14, fontweight='bold', pad=15)

# Add grid for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Remove unnecessary spines
sns.despine(left=True, bottom=True)

# Show plot
plt.show()


```

##### 1. Why did you pick the specific chart? 

A bar chart is simple and effective for comparing the most common job roles and their average salaries. It makes it easy to see which roles are most in demand and how they are paid. 

##### 2. What is/are the insight(s) found from the chart? 
The chart helps us see which job titles appear most frequently and their average salaries. This is useful for understanding which roles are in high demand and whether they pay well compared to others. It can guide job seekers on popular career choices and help companies set competitive salaries.
